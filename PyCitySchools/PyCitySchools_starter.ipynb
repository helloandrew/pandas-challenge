```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate

```python
#***why each one carries "school" in calculations?
#find the school types, school names as index
school_types = school_data.set_index(["school_name"])["type"]
school_table = pd.DataFrame(data =school_types)

#find the total students per school
students_per_school = student_data.groupby(["school_name"]).count()["Student ID"]

#find the total school budget per school
school_total_budget = school_data.groupby(["school_name"]).sum()["budget"]

#find the budget per student at each school
budget_per_student = school_total_budget / students_per_school

#find the average math score at each school
average_math_per_school = school_data_complete.groupby(["school_name"]).mean()["math_score"]

#find the average reading score at each school
average_reading_per_school = school_data_complete.groupby(['school_name']).mean()["reading_score"]

#find the percentage of passing math at each school
pass_math = school_data_complete[(school_data_complete["math_score"]>=70)]
pass_math_number = pass_math.groupby(['school_name']).count()["student_name"]
percent_math_school = pass_math_number / students_per_school * 100

#find the percentage of passing reading at each school 
pass_reading =school_data_complete[school_data_complete["reading_score"]>=70]
pass_reading_number = pass_reading.groupby(["school_name"]).count()["student_name"]
percent_reading_school = pass_reading_number / students_per_school *100

#find the Overall Passing Rate (Average of the above two)
overall_pass_rate = (percent_math_school + percent_reading_school) / 2
```

```python
#***why no "[]"??
#sometimes no [], sometimes [] and sometimes [""]
per_school_data = {"School Type":school_types,
                   "Total Students":students_per_school,
                   "Total School Budget":school_total_budget,
                   "Per Student Budget":budget_per_student,
                   "Average Math Score":average_math_per_school,
                   "Average Reading Score":average_reading_per_school,
                   "% Passing Math":percent_math_school,
                   "% Passing Reading":percent_reading_school,
                   "Overall Passing Rate":overall_pass_rate}

per_school_summary = pd.DataFrame(data = per_school_data)
per_school_summary
```

TOP PERFORMING SCHOOLS (BY PASSING RATE)

```python
#note: ascending -- small to large -- False
top_schools=per_school_summary.sort_values(["Overall Passing Rate"], ascending=False)
top_schools.head()
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
#9th grade students
nineth_grade = school_data_complete[school_data_complete["grade"] == "9th"]
nineth_grade_math = nineth_grade.groupby(["school_name"]).mean()["math_score"]

#10th grade students
tenth_grade = school_data_complete[school_data_complete["grade"]=="10th"]
tenth_grade_math = tenth_grade.groupby(["school_name"]).mean()["math_score"]
tenth_grade_math

#11th grade students
eleventh_grade = school_data_complete[school_data_complete["grade"]=="11th"]
eleventh_grade_math = eleventh_grade.groupby(["school_name"]).mean()["math_score"]

#12th grade students
twelve_grade = school_data_complete[school_data_complete["grade"]=="12th"]
twelve_grade_math = twelve_grade.groupby("school_name").mean()["math_score"]
```

```python
math_by_grade = {"9th":nineth_grade_math,
                 "10th":tenth_grade_math,
                 "11th":eleventh_grade_math,
                 '12th':twelve_grade_math}
math_by_grade_table = pd.DataFrame(data = math_by_grade)
math_by_grade_table["9th"] = math_by_grade_table["9th"].map("{:,.2f}".format)
math_by_grade_table["10th"] = math_by_grade_table["10th"].map("{:,.2f}".format)
math_by_grade_table["11th"]=math_by_grade_table["11th"].map("{:,.2f}".format)
math_by_grade_table["12th"] = math_by_grade_table["12th"].map("{:,.2f}".format)
math_by_grade_table
```
